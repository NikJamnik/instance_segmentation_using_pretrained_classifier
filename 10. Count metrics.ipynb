```
from PIL import Image
import numpy as np
import json
```

```
metric_arr = []

for temp_index in range(1, 177):
  mask_path = f"/content/drive/My Drive/Course Work/Algo 2014/images/result masks/dog_{temp_index}.png"
  ann_path = f"/content/drive/My Drive/Course Work/Algo 2014/images/validation annotations/dog_{temp_index}.json"

  mask = np.array(Image.open(mask_path))
  with open(ann_path, "r") as read_file:
    ann = json.load(read_file)

  pred_mask_area = 0
  true_mask_area = ann[2] * ann[3]
  intersection_area = 0

  for i in range(mask.shape[0]):
    for j in range(mask.shape[1]):
      if mask[i][j] > 250:
        pred_mask_area += 1
        if i >= ann[0] and i <= ann[0] + ann[2]:
          if j >= ann[1] and j <= ann[1] + ann[3]:
            intersection_area += 1

  intersection_over_union = intersection_area / (pred_mask_area + true_mask_area - intersection_area)
  metric_arr.append(intersection_over_union)
  print(f"Done {temp_index}")
```

```
m_arr = np.array(metric_arr)
```

```
print(np.round(m_arr.min(), 4), np.round(m_arr.max(), 4), np.round(m_arr.mean(), 4), np.round(m_arr.std(), 4))
```
